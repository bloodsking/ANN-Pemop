```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = data.drop('name', axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('status',axis=1)
y = data['status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
print ("Train X shape: " + str(X_train.shape))
print ("Train Y shape: " + str(y_train.shape))
print ("Test X shape: " + str(X_test.shape))
print ("Test Y shape: " + str(y_test.shape))
```

```python
# define activation function
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
# define weight (w) and bias (b)
def initialize_parameters(dim):
    w = np.random.randn(dim, 1)*0.01
    b = 0
    return w, b
```

```python
# function for: Forward and Back Propagation
def propagate(w, b, X, Y):   
    m = X.shape[1]
    
    #calculate activation function
    A = sigmoid(np.dot(w.T, X)+b)
    
    #find the cost
    cost = (-1/m) * np.sum(Y * np.log(A + 1e-15) + (1 - Y) * np.log(1 - A + 1e-15))

    #find gradient (back propagation)
    dw = (1/m) * np.dot(X, (A-Y).T)
    db = (1/m) * np.sum(A-Y)
    cost = np.squeeze(cost)
    grads = {"dw": dw,
             "db": db} 
    return grads, cost
```

```python
# function for: Optimization (Gradient Descent) for updating parameters
def gradient_descent(w, b, X, Y, iterations, learning_rate):
    costs = []
    for i in range(iterations):
        grads, cost = propagate(w, b, X, Y)
        
        #update parameters
        w = w - learning_rate * grads["dw"]
        b = b - learning_rate * grads["db"]
        costs.append(cost)
        if i % 500 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
    
    params = {"w": w,
              "b": b}    
    return params, costs
```

```python
def predict(w, b, X):    
    # number of example
    m = X.shape[1]
    y_pred = np.zeros((1,m))
    w = w.reshape(X.shape[0], 1)
    
    A = sigmoid(np.dot(w.T, X)+b)
    
    for i in range(A.shape[1]):
        y_pred[0,i] = 1 if A[0,i] >0.5 else 0 
        pass
    return y_pred
```

```python
# MAIN MODEL
def model(train_x, train_y, test_x, test_y, iterations, learning_rate):
    w, b = initialize_parameters(train_x.shape[0])
    parameters, costs = gradient_descent(w, b, train_x, train_y, iterations, learning_rate)
    
    w = parameters["w"]
    b = parameters["b"]
    
    # predict 
    train_pred_y = predict(w, b, train_x)
    test_pred_y = predict(w, b, test_x)
    print("Train Acc: {} %".format(100 - np.mean(np.abs(train_pred_y - train_y)) * 100))
    print("Test Acc: {} %".format(100 - np.mean(np.abs(test_pred_y - test_y)) * 100))
    
    return costs
```

```python
# Call the main model
iterations = 2000
learning_rate = 0.00001
costs = model(X_train.T, y_train.values.reshape(1, -1), 
              X_test.T, y_test.values.reshape(1, -1), 
              iterations, learning_rate)
```

```python
# Plotting the cost vs iteration
plt.plot(costs)
plt.ylabel('cost')
plt.xlabel('iterations')
plt.title("Learning rate = 0.00001")
plt.show()
```
